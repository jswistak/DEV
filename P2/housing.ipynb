# Data transformations

```python
from sklearn.impute import SimpleImputer
rng = np.random.RandomState(42)

def process_data(missing_rate=0.0, strategy='mean'):
    # split data into trian and test
    df_train = df.sample(frac=0.8, random_state=42)
    df_test = df.drop(df_train.index)

    # retrieve X and y
    X_train, y_train = df_train[['LSTAT', 'RM']], df_train[['MEDV']]
    X_test, y_test = df_test[['LSTAT', 'RM']], df_test[['MEDV']]


    # introduce nan to the training set
    n_samples, n_features = X_train.shape    
    n_missing_samples = int(n_samples * n_features * missing_rate)

    missing_samples = rng.randint(low=0, high=n_samples * n_features, size=n_missing_samples)
    for val in missing_samples:
        #print(val // n_features, val % n_features)
        X_train.iloc[val // n_features][val % n_features] = np.nan
    
    imp = SimpleImputer(missing_values=np.nan, strategy=strategy)
    X_train = pd.DataFrame(imp.fit_transform(X_train), columns=X_train.columns)


    X_train = X_train.to_numpy()
    y_train = y_train.to_numpy()
    X_test = X_test.to_numpy()
    y_test = y_test.to_numpy()
    
    return X_train, X_test, y_train, y_test
```

# Model building

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def linear_regression(X_train, X_test, y_train, y_test):
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    return mse
```

```python
res = pd.DataFrame(columns=['missing_rate', 'strategy', 'accuracy'])

for starategy in ['mean', 'median', 'most_frequent']:
    print("Strategy: ", starategy)
    for missing_rate in [0.0, 0.05, 0.1, 0.15]:
        print("Missing Rate: ", missing_rate)
        X_train, X_test, y_train, y_test = process_data(missing_rate=missing_rate, strategy=starategy)
        acc = linear_regression(X_train, X_test, y_train, y_test)
        res = res.append({'missing_rate': missing_rate, 'strategy': starategy, 'accuracy': acc}, ignore_index=True)
        print(acc)
```

```python
sns.lmplot(data=res, x='missing_rate', y='accuracy', hue='strategy')
```

# Random forest regression

```python
from sklearn.ensemble import RandomForestRegressor

def random_forest(X_train, X_test, y_train, y_test):
    rf = RandomForestRegressor(n_estimators=10, random_state=42)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    return mse
```

```python
res = pd.DataFrame(columns=['missing_rate', 'strategy', 'accuracy'])

for starategy in ['mean', 'median', 'most_frequent']:
    print("Strategy: ", starategy)
    for missing_rate in [0.0, 0.05, 0.1, 0.15]:
        print("Missing Rate: ", missing_rate)
        X_train, X_test, y_train, y_test = process_data(missing_rate=missing_rate, strategy=starategy)
        acc = random_forest(X_train, X_test, y_train, y_test)
        res = res.append({'missing_rate': missing_rate, 'strategy': starategy, 'accuracy': acc}, ignore_index=True)
        print(acc)
```

```python
sns.lmplot(data=res, x='missing_rate', y='accuracy', hue='strategy')
```
