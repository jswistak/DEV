# Visualization recap in Python
Import libraries for the plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load csv into dataframe


```python
df = pd.read_csv('winequality-all.csv')
df.head()
```

Investigating the distribution of alcohol variable.

I have used histplot with kde since it is the best to see the distribution of one type 

```python
# a) investigating the distribution of alcohol variable 
hist = sns.histplot(df['alcohol'], kde=True, bins=20, color='red')
hist.set_title('Alcohol Content')
hist.set_xlabel('Alcohol %')
```

Comparing the distribution of alcohol variable between two types of wine i.e. red and white 

I have used boxplot to visualize the distribution, it gave me the informations about outliers which was not possible to see on displot


```python
# b) comparing the distribution of alcohol variable between two types of wine i.e. red and white 
#hist = sns.displot(df, x='alcohol', hue='color', kde=True, bins=20)
#hist.set(title='Alcohol Content by Type', xlabel='Alcohol %')
```

```python
box = sns.boxplot(x='color', y='alcohol', data=df)
box.set(title='Alcohol Content by Type', xlabel='Type', ylabel='Alcohol %')
```

Comparing the distribution of alcohol variable in each of possible quality group defined by response variable

I have used both boxplot and violinplot since it gives me a view of distribution across different qualities


```python
# c) comparing the distribution of alcohol variable in each of possible quality group defined by response variable
box = sns.boxplot(x='response', y='alcohol', data=df)
box.set(title='Alcohol Content by Quality', xlabel='Quality', ylabel='Alcohol %')
```

```python
#sns.boxenplot(x="response", y="alcohol",
#              color="b",
#              scale="linear", data=df)
```

```python
sns.violinplot(data=df, palette="Set3", bw=.2, cut=1, linewidth=1, x="response", y="alcohol")
```

Percentage of red and white wines within each quality group.
I have used barplot to see the procentage across each quality group


```python
# d) percentage of red and white wines within each quality group 
# group by color and response and count the number of rows in each group
grouped = df.groupby(['color', 'response']).size().reset_index(name='counts')
grouped['%'] = grouped['counts'] / grouped['counts'].sum() * 100
grouped

barplot = sns.barplot(x='response', y='%', hue='color', data=grouped)
barplot.set(title='Percentage of Red and White Wines by Quality', xlabel='Quality', ylabel='%')
```

Investigating the relationship between variables describing acidity of the wines
I have used scatterplot ot see the relation between volatile and fixed acidity

```python
# e) investigating the relationship between variables describing acidity of the wines
df.head()
sns.scatterplot(x='fixed.acidity', y = 'volatile.acidity', data=df)
```
